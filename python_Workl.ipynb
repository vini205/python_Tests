```python
print('Bem vindo a Pizzaria do Vinícus')
class Pizza:
  def __init__(self,cod,desc,pM,pG):
    self.codigo= cod
    self.descricao = desc
    self.media= pM
    self.grande= pG
  def show(self):
    grandeM = float(self.grande)
    mediaM= float(self.media)
    g = format(grandeM,'.2f')
    m = format(mediaM,'.2f')
    print(f'|   {self.codigo}   | {self.descricao}  | R$ {m}    | R$ {g}     |')

napolitana= Pizza('21','Napolitana',20,26)
margherita= Pizza('22','Margherita',20,26)
calabresa= Pizza('23','Calabresa',25,32.5)
toscana=Pizza('24','Toscana',30,39)
portuguesa=Pizza('25','Potuguesa',30,39)

print('------------------------Cardápio---------------------')
print('| Código | Decrição    | Pizza Média | Pizza Grande |')
napolitana.show()
margherita.show()
calabresa.show()
toscana.show()
portuguesa.show()

i=1;
preco=0
while(i==1):
  tamanho= input('Qual é o tamanho da pizza que deseja (G/M): ')
  if(tamanho!='G' and tamanho!='M'):
    print('Opção inválida')
    continue
  sabor = input( 'Entre com o código do sabor desejado: ')
  sabor = int(sabor)
  if (sabor==21):
    print(f'Voce pediu Napolitana')
    if(tamanho=='G'):
      preco+=26
    else:
      preco+=20
  elif(sabor==22):
    print(f'Você pediu Margherita')
    if(tamanho=='G'):
      preco+=26
    else:
      preco+=20
  elif(sabor==23):
    print(f'Você pediu Calabresa')
    if(tamanho=='G'):
      preco+=32.5
    else:
      preco+=25
  elif(sabor==24):
    print(f'Você pediu Toscana')
    if(tamanho=='G'):
      preco+=39
    else:
      preco+=30
  elif(sabor==25):
    print(f'Você pediu Portuguesa')
    if(tamanho=='G'):
      preco+=39
    else:
      preco+=30
  else:
    print('Código Inválido')  
    continue
  print('Deseja pedir mais alguma coisa?')
  print('Sim -> 1')
  print('Não -> 0')
  mais= input('')
  i=int(mais)
  if(i==0):
    preco = format(preco,'.2f')
    print(f'Você gastou {preco} reais')
  else:
    continue


```
